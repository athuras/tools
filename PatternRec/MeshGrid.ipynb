```python
%pylab inline
import numpy as np
import matplotlib
import matplotlib.pyplot as plt
import pdfs
from mpl_toolkits.mplot3d import Axes3D
from matplotlib import cm
```

```python
N = pdfs.Multivariate_Gauss(2)"
```

```python
X = np.arange(-4,4,0.1)
Y = np.arange(-4,4,0.1)
X, Y = np.meshgrid(X, Y)
xn, yn = X.shape
W = X * 0
```

```python
for xk in xrange(xn):
    for yk in xrange(yn):
        W[xk, yk] = N.evaluate(np.transpose(np.matrix((X[xk, yk], Y[xk, yk]))))
```

```python
fig = plt.figure()
ax = Axes3D(fig)
ax.plot_surface(X, Y, W, rstride=1, cstride=1, cmap=cm.rainbow)
```
