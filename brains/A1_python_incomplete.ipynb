```python
%pylab --no-import-all inline
import numpy as np
import matplotlib
import matplotlib.pyplot as plt
from mpl_toolkits.mplot3d.axes3d import Axes3D
import neurons as N
import utils
from itertools import izip

def rmse(x):
    return np.sqrt(np.mean(x**2))
```

## 1 Representation of Scalars

### 1.1 Basic Encoding and Decoding

#### 1.1.1 Rectified Linear Neuron Activation and Decoding under Noise

```python
n = 16
dx = 0.05
x = np.arange(-1, 1 + dx, dx)
x_extremes = np.array([-1, 1])
x_sel = np.random.randint(0, x_extremes.size, size=n)

# The parameters we would like for our neurons
# To avoid the equations having no solutions ...
x_targets = x_extremes[x_sel] 
encoders = x_extremes[x_sel]
x_intercepts = np.random.uniform(-1, 1, size=n)
y_targets = np.random.uniform(100, 200, size=n)

# Generate the Current Models for each neuron given the fitted parameters
ltf = N.linear_tuning_fit(x_targets, x_intercepts, y_targets, encoders)
alphas, J_biases = ltf[0,:], ltf[1,:]
A = N.rect_lin(N.activity(x, encoders, alphas, J_biases))
noise_power = A.max() * 0.2
noise = np.random.normal(loc=0, scale=noise_power, size=A.shape)


f, (a1, a2) = plt.subplots(2, 1, figsize=(8, 5), sharex=True)
a1.plot(x, A)
a1.set_xlim(*x_extremes)
a1.set_title("Rectilinear Activity")
a1.set_ylabel("a [Hz]") 

a2.plot(x, np.abs(A + noise))
a2.set_ylabel("a [Hz]")
a2.set_title("With noise");
```

```python
d = N.lstsq_decode(A, x, dx=dx)
dn = N.lstsq_decode(np.abs(A + noise), x, dx=dx)

f, axes = plt.subplots(2, 2, figsize=(8, 5), sharex=True, sharey=True)
x_hat = [np.dot(a, dec)  for a in (A, A + noise) 
                         for dec in (d, dn)]

labels = ["%s,RMSE: %f" % (t, rmse(x - y)) for t, y in izip(["Clean, clean ",
                                                              "Clean, noised ",
                                                              "Noisy, clean ",
                                                              "Noisy, noised "], x_hat)]
for ax, xh, label in izip(axes.flat, x_hat, labels):
    ax.plot(x, x - xh)
    ax.set_title(label)
    ax.set_xlim(*x_extremes)
```
