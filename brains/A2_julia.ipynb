## 1.1 Gaussian White Noise

```julia
T, dt = 1., 0.001
r = 0.5
bw = [5., 10., 20.]

# Used to simplify plotting
function show_signal(T, dt, r, mask, bw)
    f, axes = subplots(length(bw), 1, figsize=(length(bw) * 2, 5), sharex=true)
    t = linspace(0, T, int(T/dt))
    for (a, b) in zip(axes, bw)
        x = gen_wgn_signal(T, dt, r, mask, b)
        a[:plot](t, x)
        a[:set_title]("Bandwidth: $(b) [Hz]")
    end
    return f
end
```

### 1.1 a)
The highest frequency components of the fourier spectra of the signal shown is approximated by the number of 'peaks' in the signal. 
In this sense, when using an ideal low-pass filter, we would not expect to see more then 5-peaks for the 5 Hz bandwidth signal. 
This pattern is repeated for all bandwidths plotted below.

```julia
f1 = show_signal(T, dt, r, ideal_LPF, bw)
```

### 1.1 b) 
The mean coefficient magnitude of the ideal low-pass-filtered signal is around 0.025. In isolation this isn't immediately valuable information, however in comparison to the chart in 1.2b), it is shown that this value is slightly lower than the equivalent bandwidth gaussian filter. 
This is a direct result of zero-ing the coefficients greater than the bandwidth, leading to 0-magnitude coefficients (related to frequencies greater than the bandwidth) to pull down the mean.

```julia
# Now Showing the coefficients
function coeff_norm(mask, trials)
    norms = [
        begin
            freqs, coeffs = wgn_coeffs(T, dt)
        masked = mask_coeffs(freqs, coeffs, mask, 10)
        mean(abs(masked))::Float64
        end
        for i in 1:trials]
    f, (a1, a2) = subplots(2, 1)
    a1[:plot](norms)
    a1[:set_title]("Mean Fourier Coefficient Magnitude")
    a1[:set_xlabel]("Trial")
    a1[:set_ylabel]("Mean Magnitude")
    a2[:hist](norms, bins=15)
    a2[:set_title]("Mean Fourier Coefficient Magnitude Histogram")
    return f
end
coeff_norm(ideal_LPF, 100)
```

*On Performance*:
The most noticeable problem with this simulation is caused by round-off error associated with calculating the number of simulation steps that (roughly) equate to the refractory time. 
As shown in 2a, round-off error causes the simulation of the neuron with a peak firing rate of 150Hz to artificially increase its activity by 10%.

The brute force solution is to simply increase the resolution of the simulation, but ultimately this is expensive.
My solution however operates only on the recovery steps (time steps within refractory period), and involves varying the number of time steps to wait, this involves intelligently over-, and under-estimating the refractory period, so that the mean simulated refractory time is as close as possible to the true refractory period.

Computationally, this adds trivial complexity (essentially constant proportional to the number of neurons) while (in the 2a/150Hz case) offering roughly 10% increased spike-rate fidelity.

## 4: Computing an Optimal Filter
My implementation is designed to closely match the NumPy example. This is wrapped into a function later for convenience.

```julia
function conv_same(a, b) 
    n, m = length(a), length(b)
    z = max(m, n)
    left = div(z, 2)
    right = z - left
    center = div(n + m - 1, 2)
    return conv(a, b)[center - left + 1: center + right]
end

T = 4.
dt = 0.001
desired_rms, limit = 0.5, 5.

# Break up the gen_wgn_signal into its pieces to split out the coeffs
# My function only returns the real signal, not the coefficients, so 
# What follows is essentially the step-wise re-implementation of 
# gen_wgn_signal(...).
x = gen_wgn_signal(T, dt, 0.5, ideal_LPF, 5)
X = fftshift(fft(x))

Nt = length(x) # The number of simulation time-steps
t = [0:Nt-1] .* dt # The time (in seconds).

# Neuron Parameters
t_ref, t_rc = 0.002, 0.02  # The transient properties of the neurons
x0, x1 = 0., 1.
a0, a1 = 1., 150.

# Fit the alpha, and Jbias parameters given 'arbitrary' points (vs. intercepts).
eps = t_rc / t_ref
r1, r2 = 1. / (t_ref .* [a0 a1])
f1, f2 = ([r1 r2] .- 1.) ./ eps
alpha = (1. / exp(f2) - 1.) - 1. / (exp(f1) - 1.)/(x1 - x0)
x_threshold = x0 - 1. / (alpha * (exp(f1) - 1.))
Jbias = 1 - alpha * x_threshold

# Simulate the two neurons
spikes = float64(lif_sim(x, dt, [1., -1.], [params; params], 
                 t_ref=t_ref, t_rc=t_rc) .== 1.)

omega = 2pi * ([1:Nt] / T .- Nt./(2.T))  # Frequencies for constructing the Gaussian

r = vec(spikes[1,:] .- spikes[2,:])   # The spike signal
R = fftshift(fft(r))                 # The fourier spectra of the Real-signal R (half-space)

sigma_t = 0.025                     # The Gaussian envelope std-dev
W2 = complex(exp(-(omega.*omega) * sigma_t^2)) # The Gaussian Kernel/window (in freq-domain)
CP = X .* conj(R)                   # Spike-Stimulus correlate power spectrum
WCP = conv_same(CP, W2)             # smoothed (above)
RP = R .* conj(R)                   # Power spectrum of spike signal
WRP = conv_same(RP, W2)             # Smoothed spike power spectrum
XP = X .* conj(X)                   # Power spectrum of signal x
WXP = conv_same(XP, W2)             # Smoothed stimulus power spectrum

shifting(x) = fftshift(real(ifft(ifftshift(x))))  # Shenanigans

H = WCP ./ WRP            # The optimal filter (power)
h = shifting(H)           # The optimal filter (as a real signal)


XHAT = H .* R             # The optimally filtered signal (in F-space)
xhat = shifting(XHAT)     # as a real signal
0; # Silence output
```

```julia
# Plot stuff
set_labels(a, x, y) = begin a[:set_xlabel](x); a[:set_ylabel](y) end
freq = omega ./ (2pi)

f, (a1, a2) = subplots(1, 2, figsize=(12, 6))
a1[:plot](freq, sqrt(XP), label="Stimulus X Energy Spectrum")  # Stimulus Power-Spectrum
a1[:legend](loc="best")
set_labels(a1, "Frequency [Hz]", "|X(\$\\omega\$)|")

a2[:plot](freq, sqrt(RP), label="Spike Signal R Energy Spectrum")  # Spike Signal Power-Spectrum
a2[:legend](loc="best")
set_labels(a2, "Freqency [Hz]", "|R(\$\\omega\$)|")

f2, (a1, a2) = subplots(1, 2, figsize=(12, 6))
a1[:plot](omega ./ (2pi), real(H))  # The fourier spectrum of the filter
a1[:legend](loc="best")
a1[:set_xlim](-50, 50)
set_labels(a1, "Frequency [Hz]", "|H(\$\\omega\$)|]")

a2[:plot](t .- T/2, h)  # the filter ... the "T/2" term is to center it (it is acausal)
a2[:set_title]("Optimal Filter")
a2[:set_xlim](-.5, .5)
set_labels(a2, "Time", "h(t)")

f3, ax = subplots(1, 1, figsize=(10, 4))
ax[:plot](t, r, label="Spikes, r", alpha=0.2)
ax[:plot](t, x, linewidth="2", label="stimulus, x")
ax[:plot](t, xhat, label="Filtered Spikes")
ax[:set_title]("Optimal Filter Decoding")
ax[:legend](loc="best")
ax[:set_xlabel]("Time [s]")
```

```julia
shifting(x) = fftshift(real(ifft(ifftshift(x))))  # Shenanigans
function optimal_filters(enc, params, T, dt, mask, limit, desired_rms=0.5)
    x = gen_wgn_signal(T, dt, desired_rms, mask, limit)
    X = fftshift(fft(x))

    Nt = length(x) # The number of simulation time-steps
    t = [1:Nt] .* dt
    
    # Neuron Parameters
    t_ref, t_rc = 0.002, 0.02
    x0, x1 = 0., 1.
    a0, a1 = 1., 150.

    spikes = float64(lif_sim(x, dt, enc, params, 
                 t_ref=t_ref, t_rc=t_rc) .== 1.)

    r = vec(spikes[1,:] .- spikes[2,:])   
    R = fftshift(fft(r))                 

    sigma_t, omega = 0.025, 2pi * ([1:Nt] / T .- Nt./(2.T))
    W2 = complex(exp(-(omega.*omega) * sigma_t^2))
                
    WCP = conv_same(X .* conj(R), W2)           
    WRP = conv_same(R .* conj(R), W2)           
    H = WCP ./ WRP      
    h = shifting(H)        
    
    XHAT = H .* R
    xhat = shifting(XHAT)  
    return t, omega, h, H
end
```

```julia
bw = [2, 10, 30.]
q = [optimal_filters([1., -1.], [params; params], T, dt, ideal_LPF, l) 
    for l in bw]
f, axes = subplots(3, 1, figsize=(13, 6), sharex=true)
for (i, a) in enumerate(axes)
    (ti, om, hh, HH) = q[i]
    a[:plot](ti .- T/2, hh)
    a[:set_title](@sprintf "limit: %i [Hz]" bw[i])
    a[:set_xlim](-0.5, 0.5)
end
```

### 4e)
**Above**: as the bandwidth is increased, the filter becomes narrower, and contains more oscillations.
These are both results of having more of the filter's energy located in higher-frequencies.
Intuitively, in order to not attenuate the higher frequencies in the spike signal, the filter must correspondingly become higher-frequency.

### 4f)
**Above**: As the duration of the WGN stimulus increases (same bandwidth) the optimal filter shape/spectrum does not noticeably change (as long as the bandwidth of the WGN signal is smaller than the Nyquist frequency).
This is because the WGN stimulus has a well-defined fourier spectra, which is independent of the duration used for the simulations.
Similarly, the fourier spectra of the spike response is relatively invariant given a particular WGN spectra.

## 5: Using Post Synaptic Currents as a Filter

```julia
psc_filter(t, n, tau) = (z = t.^n .* exp(-t ./ tau); z ./ sum(z))

t = [1:500] .* 0.0001
f, a = subplots(1, 1)
map([0, 1, 2]) do q # more fun Julia syntax!
    a[:plot](psc_filter(t, q, 0.007), label=@sprintf "n = %i" q)
end
a[:legend](loc="best")
title("Post-Synaptic Current Filter")
xlabel("t, [ms]")
ylabel("h(t)")
```

### 5 a)
**Above**: Changes to n (the 'shape' parameter of the distribution) leads to both a shifting the distributions centroid further back in time, as well as severely attenuating high-frequency components (the 'spike' gets smoothed).
This essentially performs a wider-band low-pass-filter (as n-increase).
$\hat{x}$ will thus become smoother (less influence of noise), but also more delayed, that is to say, the filtered waveform will become increasingly out-of-phase with the stimulus as n increases.

```julia
f, a = subplots(1, 1)
map([0.002, 0.005, 0.01, 0.02]) do q # more fun Julia syntax!
    a[:plot](psc_filter(t, 0., q), label=@sprintf "\$\\tau\$ = %.3f" q)
end
a[:legend](loc="best")
title("Post-Synaptic Current Filter")
xlabel("t, [ms]")
ylabel("h(t)")
```

### 5b)
**Above**: Increasing the time-constant (or 'scale' parameter) of the filter shifts the distribution's centroid farther back in time. Thus the filtered waveform will become more out of phase as $\tau$ increases. Additionally, **decreasing** $\tau$ increases the bandwidth of the filter, implying the filtered waveform will contain additional high-frequency components which may, or may not be desireable.

## Signals
WGN generators and such.

```julia
module Signals

import Kernels: ideal_LPF

export rms, rescale_rms, wgn_coeffs, mask_coeffs, gen_wgn_signal

rms(x) = sqrt((1. / prod(size(x))) * sum(x .* x))
rescale_rms(x, z::Real) = (z / rms(x)) .* x

# generate a colored noise signal, whose color is based on the application
# of a DFT coefficient 'mask_fn'. For examples of mask_fns, see module
# Kernels
function gen_wgn_signal(T::Real, dt::Real, rms_power::Real,
                        mask_fn::Function=ideal_LPF,
                        mask_params...)
    const n = int(T / dt)
    freqs, coeffs = wgn_coeffs(T, dt)
    coeffs = mask_coeffs(freqs, coeffs, mask_fn, mask_params...)
    x = irfft(coeffs, n)  # Shenanigans with the size ...
    return rescale_rms(x, rms_power)
end

# Return the frequencies, and fourier coefficients of white noise spectrum
# of a Real signal of period T, sample_period dt.
function wgn_coeffs(T::Real, dt::Real)
    const n = int(T / dt)
    const n_half = div(n, 2)
    freqs = 2pi/dt .* [i / n for i in 0:n_half]
    coeffs = complex(randn(n_half + 1), randn(n_half + 1))
    coeffs[1] = 0  # No DC component
    return freqs, coeffs
end

# Apply the mask function to the DFT coefficients coeffs.
function mask_coeffs(freqs, coeffs, mask_fn::Function, mask_params...)
    mask = mask_fn(freqs, mask_params...)
    return coeffs .* mask
end
end
```

## Kernels
Discrete Filters for use with other modules

```julia
module Kernels
export ideal_LPF, gaussian_LPF

# Returns multiplicative mask. Suitable for application on
# a half-fourier coefficient array.
function ideal_LPF(freqs, cutoff)
    const sel = freqs .> 2pi*cutoff
    k = ones(size(freqs))
    k[sel] = 0
    return k
end

# elementwise gaussian kernel evaluation on frequency coeffs
function gaussian_LPF(freqs, bandwidth)
    const cutoff = 2pi*bandwidth
    k = exp(-(freqs .* freqs) ./ (cutoff * cutoff))
    return k
end
end
```

## Neurons
The "Main Event"

```julia
module Neurons

using utils, tuning

export activity, ddot, decode, lif_fit, lin_fit
export lif_sim, current, lif_delta


activity(x, enc, p) = x * enc' .* p[:,1]' .+ p[:,2]'
ddot(x, y, k) = sum(x .* y, k)

# where noise must be scalar, stationary independent etc.
function decode(A, x, dx, noise=0.)
    Gamma = A' * A * dx
    Gamma += diagm(repmat([noise], size(Gamma)[1]))
    Upsilon = A' * x * dx
    return Gamma \ Upsilon
end


# Fit alpha, and bias terms for so-called 'leaky integrate and fire'
# tuning curves.
function lif_fit(x_max, x_int, y_max, enc, t_ref=0.002, t_rc=0.02)
    const B = exp((1. / t_rc) .* (t_ref - 1. / y_max))
    alpha = (1 ./ (1. - B) - 1) ./ ddot(x_max - x_int, enc, 2)
    bias = 1. - alpha .* ddot(x_int, enc, 2)
    return cat(2, alpha, bias)
end

# Calculate Params
function dual_point_lif_fit(x, a; t_ref=0.002, t_rc=0.02)
    const eps = t_rc / t_ref
    r = 1. / (t_ref .* a)
    f = (r .- 1) ./ eps
    alpha = (1. / exp(f[2]) - 1) - 1. / (exp(f[1]) - 1) ./ (x[2] - x[1])
    x_threshold = x[1] - 1 ./ (alpha * exp(f[2]) - 1)
    Jbias = 1 - alpha .* x_threshold
    return [alpha Jbias]
end

# linear tuning fit
# returns an array of alpha, bias terms, one for each of x_max
function lin_fit(x_max, x_int, y_max, enc)
    alpha = y_max ./ ddot((x_max - x_int), enc, 2)
    bias = - ddot(x_int, enc, 2) .* alpha
    return cat(2, alpha, bias)
end

# Possibly the same as 'activity', I try not to think about it...
function current(x::Array{Float64}, encoders::Array{Float64}, params::Array{Float64})
    return params[:,1] .* (encoders * x') .+ params[:,2]
end

lif_delta(J::Array{Float64}, V::Array{Float64}, t_rc=0.002) = (J .- V) ./ t_rc

# Type Annotations all over the place, mainly because when I was debugging shit
# and types were going crazy
# If you want to do this with abstract types (i.e Real) use at own risk.
function lif_sim(x::Array{Float64},
                 dt::Float64,
                 encoders::Array{Float64},
                 params::Array{Float64};
                 t_ref=0.002, t_rc=0.02, spike_ref=1.)
    const k = size(x, 1)
    const n = size(params, 1)
    refractory_steps = int(t_ref / dt)  # BEWARE roundoff error

    V_sim = Array(Float64, n, k)
    V_sim[:,1] = 0.  # Initial state of zero
    J = current(x, encoders, params)::Array{Float64}
    
    # if positive, this is the number of steps a neuron must wait
    # before it can fire again.
    ref_state = zeros(Int64, n)

    # i is the index corresponding to the i-th time step.
    for i in 2:k
        dv_dt = lif_delta(J[:, i], V_sim[:, i - 1], t_rc) .* dt
        V_sim[:, i] = V_sim[:, i - 1] .+ dv_dt  # Accumulate, business as usual
        V_sim[ref_state .> 0, i] = 0.           # zero the recovering neurons
        
        spikes = V_sim[:, i] .> 1.
        V_sim[spikes, i] = spike_ref            # Set the spikes to the reference value
        V_sim[:, i] = max(V_sim[:, i], 0)       # because shenanigans happened before ...
        ref_state[spikes] = refractory_steps + 1   # Set the spike counters
        ref_state -= 1
    end
    return V_sim  # The voltage 'chart' one row-per-neuron.
end
end
```
