## Low Pass Filters

Wherein we make an ideal low-pass filter, see why that behaviour isn't so pretty (even on lena), and then start experimenting with smoother cutoffs.

```python
def ideal_mask_lp_filter(shape, cutoff):
    x = np.linspace(-shape[0] / 2, shape[0] / 2., shape[0])
    y = np.linspace(-shape[1] / 2, shape[1] / 2., shape[1])
    X, Y = np.meshgrid(x, y)
    Z = np.hypot(X, Y)
    Z.reshape(shape)
    Z[Z < cutoff] = 1.
    Z[Z >= cutoff] = 0.
    return Z

def gaussian_lp_filter(shape, s):
    '''Gaussian Kernel with mean=0 defined across shape'''
    x = np.linspace(-shape[0] / 2, shape[0] / 2., shape[0])
    y = np.linspace(-shape[1] / 2, shape[1] / 2., shape[1])
    X, Y = np.meshgrid(x, y)
    Z = np.hypot(X, Y)
    Z.reshape(shape)
    Z = (1. / (2 * np.pi) ** 0.5) * np.exp(-0.5 * (Z / s)**2)
    Z /= Z.max()
    return Z
    
```

## Filter Design
Wherein we note the obnoxious sinusoidal noise in the test image.

```python
a = fft2(f_noise)
fs = np.abs(fftshift(a))
f, (a1, a2) = plt.subplots(1, 2, figsize=(6, 3), sharey=True)
a1.imshow(f_noise, **imshow_params)
a2.imshow(np.log(np.abs(fs)), cmap='gray')
a1.set_title('Original')
a2.set_title('FFT')
FIGURES.append(f)
```

Note the 2 pairs of peaks corresponding to the two off-axis sinusoidal elements added to the image. Lets try to surgically remove them with what is effectively a set of ideal band-stop filters (one for each pair).

First we have to find out at which frequencies the peaks occurr, so we'll find the indices in the FFT of each of the peaks. 

By inspection, we can see that the two sinusoidal additions do not vary spatially, so we can simplify our problem to looking for components independently in each direction! Once we've found the frequencies, its not too challenging to simply identify the spikes, compose a point/band-stop filter and apply it. Admittedly, this procedure relies on the assumption that the frequency noise was added in the laziest manner possible—which I believe is valid.

```python
def trans(x):
    return np.log(np.abs(fftshift(x)))

x_fft = fft(f_noise.flat[:256])
y_fft = fft(f_noise.T.flat[:256])
f, ax = plt.subplots(1, 1, figsize=(6, 6))
a = np.arange(-128, 128, 1)
ax.plot(a, trans(x_fft), c='r', label='X-dimension')
ax.plot(a, trans(y_fft), c='b', label='Y-dimension')
ax.grid(True)
ax.set_title('Component Analysis of f-noisy FFT')
ax.legend()
FIGURES.append(f)
```

```python
qx = argpartsort(-trans(x_fft), 5)[:5]
trans(x_fft)[qx]
qy = argpartsort(-trans(y_fft), 9)[:9]
trans(y_fft)[qy]
```

```python
points = [(64, 64), (192, 192),
          (118, 104), (138, 152)]
master_mask = np.ones_like(f_noise)
for x, y in points:
    master_mask[x, y] = 0.
```

```python
a = fft2(f_noise)
pm_denoised = fftshift(a) * master_mask
post_op = np.abs(ifft2(ifftshift(pm_denoised)))
```

```python
f, (a1, a2) = plt.subplots(1, 2, figsize=(6, 3), sharey=True)
a2.imshow(post_op, cmap='gray')
a2.set_title('Post Op')
a1.imshow(f_noise, cmap='gray')
a1.set_title('Pre Op')
FIGURES.append(f)
```
